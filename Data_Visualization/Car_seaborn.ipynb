```python
import pandas as pd
import seaborn as sns
```

```python
car = pd.read_csv('Car_sales.csv')
```

```python
car
```

```python
#NaNs will lead to this error - Chart cannot be created
sns.distplot(car['Fuel efficiency'])
```

```python
car.isnull().sum()
```

```python
sns.distplot(car['Sales in thousands'], kde=False)
```

```python
sns.regplot(data = car, y = 'Fuel efficiency', x = 'Fuel capacity', fit_reg=False)
```

```python
sns.lmplot(data = car , y = 'Fuel efficiency', x = 'Fuel capacity', fit_reg=False, hue='Vehicle type')
```

```python
sns.lmplot(data = car , y = 'Fuel efficiency', x = 'Fuel capacity', fit_reg=False, col='Vehicle type')
```

```python
sns.jointplot(data = car, x = 'Fuel capacity', y = 'Fuel efficiency', kind='kde' )
```

```python
import matplotlib.pyplot as plt
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
sns.countplot(data =car, x = 'Vehicle type')
```

```python
sns.barplot(y = '4-year resale value', x = 'Vehicle type', data = car)
```

```python
sns.stripplot(y = '4-year resale value', x = 'Vehicle type', data = car)
```

```python
sns.pairplot(car[['Horsepower', 'Fuel efficiency', 'Fuel capacity', 'Curb weight']])
```

```python
sns.heatmap(car[['Horsepower', 'Curb weight', 'Fuel capacity', 'Fuel efficiency']].corr(), annot=True)
```

```python
car[['Horsepower', 'Curb weight', 'Fuel capacity', 'Fuel efficiency']].corr()
```

```python
#Box and Whiskers plot

sns.boxplot(data = car, x = 'Vehicle type', y = 'Fuel capacity')
```

```python
sns.violinplot(data = car, x = 'Vehicle type', y = 'Fuel capacity')
```
